# IQR Outlier Removal:

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

iqr_df = df[~((df< (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

# Train and Test Split:

```python
x=df_new.drop("Happiness Score",axis=1)
y=df_new["Happiness Score"]
```

# Finding the Best Random State:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
mod=LinearRegression()
maxAccu1=0
maxRS1=0
maxAccu2=0
maxRS2=0
for i in range(0,200):
    x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.20, random_state=i)
    mod.fit(x_train,y_train)
    pred_train=mod.predict(x_train)
    pred_test=mod.predict(x_test)
    acc1=r2_score(y_train,pred_train)
    acc2=r2_score(y_test,pred_test)
    print(f"At Random state {i}, the training accuracy is: {acc1}")
    print(f"At Random state {i}, the testing accuracy is: {acc2}")
    print("\n")
    if acc1>maxAccu1:
        maxAccu1=acc1
        maxRS1=i
        
    if acc2>maxAccu2:
        maxAccu2=acc2
        maxRS2=i

print("Best train accuracy is",maxAccu1,"on Random_state",maxRS1)
print("\n")
print("Best test accuracy is",maxAccu2,"on Random_state",maxRS2)
```

Best train accuracy is 0.9995683366029277 on Random_state 78
